```python
import pandas as pd
```

```python
npr = pd.read_csv('https://raw.githubusercontent.com/yeesem/NLP_Dataset/main/npr.csv',engine = 'python',error_bad_lines = False)
```

```python
npr.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_df = 0.95,min_df = 2,stop_words = 'english')
```

```python
dtm = tfidf.fit_transform(npr['Article'])
```

```python
dtm
```

```python
from sklearn.decomposition import NMF
```

```python
nmf_model = NMF(n_components = 7,random_state = 42)
```

```python
nmf_model.fit(dtm)
```

```python
tfidf.get_feature_names_out()[200]
```

```python
len(tfidf.get_feature_names_out())
```

```python
nmf_model.components_
```

```python
nmf_model.components_.shape
```

```python
for index,topic in enumerate(nmf_model.components_):
  print(f"THE TOP 15 WORD FOR TOPIC #{index}")
  print([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-15:]])
  print('\n\n')
```

```python
topic_results = nmf_model.transform(dtm)
```

```python
topic_results
```

```python
topic_results[0]
```

```python
#Index for topic position
topic_results[0].argmax()
```

```python
topic_results.argmax(axis = 1)
```

```python
npr['Topic'] = topic_results.argmax(axis = 1)
```

```python
#Running time is greater than the LDA
#Assignment of topic is based on the domain knowledge
topic_dict = {0:'Election',1:'Music',2:'Health',3:'Legis',4:'Politics',5:'Politics',6:'Education'}
npr['Topic Label'] = npr['Topic'].map(topic_dict)
```
